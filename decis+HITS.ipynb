```python
import numpy as np

#input
n=int(input("Enter the no. of nodes in the graph : "))
gr=[]
print("Enter the matrix : ")
for i in range(n*n):
    gr.append(int(input()))

a=np.array(gr)
A=a.reshape(n,n)

print("\nMatrix formed :\n",A)

At=np.transpose(A)

print("\n\n u--> HUB WEIGHT VECTOR.......(u=A*v)\n v--> AUTHORITY WEIGHT VECTOR........(v=At*u)\n")
print()
print("Assume u=[1 1 1 1]\n\n")

print("Transpose of graph :\n",At)

v=[0 for i in range(n)]
u=[0 for i in range(n)]\

#calculate 'v'
for i in range(n):
    for j in range(n):
        if(At[i][j]==1):
            v[i]=v[i]+1           
v=np.array(v)    
print("\n\nv ==> ",v)

#calculate 'u'
for i in range(n):
    for j in range(n):
        if(A[i][j]==1):
            u[i]=u[i]+v[j]            
u=np.array(u)    
print("\n\nu ==> ",u)


print()

#print result
u=list(u)
v=list(v)
nu=[]
nv=[]
for i in range(n):
    nu.append(u[i])
    nv.append(v[i])
    
print("\n\nAUTHORITY STATUS :\n")
u.sort(reverse=True)
v.sort(reverse=True)
for i in u:
    print("N",nu.index(i)+1," ==> ",i)

print("\n\nHUB PRIORITY :\n")
for i in v:
    print("N",nv.index(i)+1," ==> ",i)
```
